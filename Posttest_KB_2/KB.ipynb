• Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.\
• Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.\

• Ubah nilai salah satu attribute bertipe angka menjadi desimal 3 angka dibelakang koma jika tidak ada, lakukan sebaliknya (dari desimal menjadi bilangan bulat).\
• Tampilkan 10 record pertama dari attribute bertipe angka.\
• Simpan dataset dengan attribute bertipe angka ke dalam file CSV.

```python
import pandas as pd
import numpy as np

from prettytable import PrettyTable

df = pd.read_csv("Airline_Passenger_Satisfaction.csv")

df.info()
```

**Menampilkan jumlah record, Nama Attribut, Jumlah Attribut dengan Tipe Objek, Jumlah Attribute dengan Tipe Angka, Jumlah Attribute dengan Nilai NULL dan memasukkannya ke dictionary**

Untuk menampilkan jumlah record yang merupakan jumlah baris dan kolom, maka akan menggunakan perintah `df.shape`.

Untuk menampilkan nama attribute dari setiap kolom, maka akan menggunakan perintah `df.columns.values`.

Untuk menampilkan jumlah Attribute yang menggunakan Tipe Objek, maka akan menggunakan perintah `df.select_dtypes(["object"]).columns`.

Untuk menampilkan Jumlah Attribute yang menggunakan Tipe Angka, maka akan menggunakan perintah `df.select_dtypes(["int64", "float64"]).columns)`.

Untuk menampilkan jumlah Attribute yang terdapat Nilai NULL, maka akan menggunakan perintah `df.loc[:, df.isnull().any()].columns`

Dan langkah selanjutnya adalah memasukkan ke dalam dictionary.

```python

informasi = {
    "Record"               : df.shape[0],
    "Attribute"            : list(df.columns.values),
    "Attribute Tipe Objek" : list(df.select_dtypes(["object"]).columns),
    "Attribute Tipe Angka" : list(df.select_dtypes(["int64", "float64"]).columns),
    "Attribute Null"       : list(df.loc[:, df.isnull().any()].columns),
}

keys = list(informasi.keys())

# Pembuatan tabel dengan prettytable
tabel_informasi = PrettyTable()
tabel_informasi.field_names = ["Informasi", "Attribute", "Jumlah"]

# Input informasi di tabel informasi
tabel_informasi.add_row([keys[0], "-", informasi[keys[0]]])

for i in range(1, 5):
    tabel_informasi.add_row([keys[i], ", \n".join(informasi[keys[i]]), len(informasi[keys[i]])])
    
print(tabel_informasi)
```

**Menampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.**

Untuk mampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka, maka kita perlu menggunakan `df.describe()`.

Untuk menampilkan nilai korelasi attribute bertipe angka, maka diperlukan perintah `df.corr()`.

**Mengubah nilai salah satu attribute bertipe angka menjadi desimal 3 angka dibelakang koma**

untuk mengubah nilai salah satu attribute bertipe angka menjadi desimal 3 angka dibelakang koma, maka di perlukan perintah `df["nama kolom"].astype("tipe data")`

```python
df["Age"] = df["Age"].astype("float64")

df['Age'].dtype
```

**note**

float di sini hanya akan menampilkan 1 digit angka di belakang koma, maka diperlukan perintah `df["nama kolom"] = df["nama kolom"].map('{:.3f}'.format)`

```python
df["Age"] = df["Age"].map('{:.3f}'.format)
df
```

**Menampilkan 10 record pertama dari attribut bertipe angka**

Untuk menampilkan 10 record pertama dari attribut bertipe angka, maka diperlukan perintah `df.head()`

```python
df_nilai_angka = df[informasi[keys[3]]]
df_nilai_angka.head(10)
```

**Menyimpan dataset dengan attribut bertipe angka ke dalam file .csv**

Untuk menyimpan dataset dengan attribut bertipe angka ke dalam file .csv, maka digunakan perintah `df.to_csv ("nama_yang_ingin_dipakai")`.
