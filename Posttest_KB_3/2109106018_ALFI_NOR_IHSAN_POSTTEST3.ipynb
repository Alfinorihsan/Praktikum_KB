```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Airline_Passenger_Satisfaction.csv")
df.info()
```

```python
x = list(df["Type of Travel"])
y = list(df["Age"])

plt.figure(figsize = (5,5))
plt.bar(x, y, label = "cutomer type", color = "red")
plt.title("rata-rata umur")
plt.legend()
plt.show()
```

rata-rata umur penumpang di penerbangan bisnis dan penerbangan personal

```python
genres = df.dropna(subset=['Customer Type']).sort_values(by=['Customer Type'])['Customer Type'].unique()

total = df.groupby(['Customer Type'])['Customer Type'].count()

patches, texts, pcts = plt.pie(total, labels=genres, autopct='%.1f%%', wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},shadow = True, textprops={'size': 'x-large'})

plt.title('Perbandingan Customer')
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

perbandingan costumer yang loyal dan tidak loyal dengan penerbangan ini yang di dapat dari seluruh data yang ada
